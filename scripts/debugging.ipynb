```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt

from reimpl_a_gn.threed.random import piecewise_uniform


def plot_piecewise_uniform_samples():
    key = jax.random.key(7)
    interval_bounds = jnp.array(
        [[-10.0, -5.0, -3.0, 5.0, 7.5], [-1.0, 0.0, 0.3, 2.0, 3.0]]
    )
    pdf_values = jnp.array(
        [[0.1, 0.025, 0.025, 0.1], [0.6216696, 0.00888099, 0.16873889, 0.08880994]]
    )
    samples = piecewise_uniform(
        key,
        intervals=interval_bounds,
        pdf_values=pdf_values,
        sample_count_per_distribution=10000,
    )
    plt.scatter(samples[0], samples[1])
    plt.show()


plot_piecewise_uniform_samples()
```
